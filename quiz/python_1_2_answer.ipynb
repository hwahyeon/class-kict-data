```python
file = open('../data/Abc1115.csv','r')
```

```python
lines = file.readlines()
file.close()
```

```python
print(lines[0:5])
```

```python
for idx,line in enumerate(lines):
    lines[idx] = line[0:-1]
    pass
print(lines[0:5])
```

```python
students = []
for line in lines:
    students.append(line.split(","))
    pass
```

```python
students[0:5]
```

```python
for student in students:
    for i in range(2,8):
        student[i] = int(student[i])
        pass
    pass
students[0:5]
```

```python
for student in students:
    avg = (student[2]+student[3]+student[4]+student[5]+student[6])/5
    student.insert(8,avg)
    pass
students[0:5]
```

```python
header = ['학번','이메일','국어','영어','수학','과학','역사','총점','평균','담임코드','지역코드','성취도코드']
```

```python
import pandas as pd
df = pd.DataFrame(students,columns=header)
```

```python
df.head()
```

```python
df.sort_values(by=['총점','학번'],ascending=[False,True]).head(15)
```
